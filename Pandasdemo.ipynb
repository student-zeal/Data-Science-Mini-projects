# Handling missing Data,fillna,dropna,interpolate

```python
new_df=df.fillna(0)# replace NaN with 0
new_df
```

```python
new_df=df.fillna(method="ffill")# replace NaN with values which carries forward (copy) the previous value
new_df # means whereever there is 0 in the table the value of the previous one shall be copied by it to that place
```

```python
new_df=df.fillna(method="bfill")
new_df
```
