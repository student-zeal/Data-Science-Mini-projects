# Axes,labels,legend,grid

```python
# same chart 3 different lines can be plotted
days=[1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [42,44,47,42,45,50,49]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best",shadow=True,fontsize="large") # to give name to the particular line there are many parameters in it
plt.grid()
```

# Bar charts

```python
company = ['GGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
```

# HIstograms

```python
# histogram is  bar chart where y axis conatins the frequency of the variable.Each range is called a bin and there are bydefault 10 ranges
blood_sugar = [113,85,90,150,88,93,115,135,80,77,129]
plt.hist(blood_sugar)
```

# NOTES
# Historgram need single Dimension array
# By default it will create/plot 10 Bins
# BIN is bucket which are ranges 
# Here we need 3 ranges or bins  

```python
plt.hist(blood_sugar,bins=3,rwidth=0.95)
```

```python
# customize the range/bin parameter
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g')
```

```python
# Let's say,we have 2 samples
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([blood_sugar_men,blood_sugar_women,],bins=[80,100,125,150],rwidth=0.95,color=['green','yellow'],label=['men','women'])
plt.xlabel("blood Sugar Range")
plt.ylabel("Total number of patients")
plt.xlabel("blood Sugar Analysis")
plt.legend()
```

# PIE CHART

```python
# how much % or pie ,individual part is contributing out of total value.

Example:
            home rent- 1400
            Food     - 600
phone/Internet bill  - 300
                car  - 410
    Other utilities  - 250
            Total    - 2960
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home rent","food","Phone/internet bills","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
# Radius
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=0.5)
plt.show()
```

```python
# showing the %

plt.pie(exp_vals,labels = exp_labels,radius=1.2,autopct='% 0.1f %%',shadow = True)
```

```python
# To explode or remove some part and also giving an angle

plt.pie(exp_vals,labels = exp_labels,radius=1.2,autopct='% 0.1f %%',shadow = True,explode=[0,0.1,0.1,0,0],startangle=45)
# SAVE CHART TO FILE USING SAVEFIG
# place the above code in place of plt.show 
plt.savefig("zealpiechart.png",bbox_inches= "tight",pad_inches=2,transparent=True )
```
