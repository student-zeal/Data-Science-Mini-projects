```python
# vertical Barplot
sns.set_context('paper')
# load dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# create the plot 
sns.barplot(x='embark_town',y='age',data=titanic,palette='PuRd',ci=None)
plt.legend()
plt.show()
```

```python
# horizontal Barplot
sns.barplot(x='age',y='embark_town',data=titanic,palette='PuRd',ci=None,orient='h')
plt.legend()
plt.show()
```

```python
# count plot
# The count plot can be thought of as a histogram across a category
titanic = sns.load_dataset('titanic')

# create the plot
sns.countplot(x='class',hue='who',data=titanic,palette='magma')
plt.title('Survivors')
plt.show()

```

# heat map

```python
# load dataset
flights=sns.load_dataset('flights')
flights.head()
```

```python
# flights = flights.pivot("month","year","passengers")
flights_wide = flights.pivot(index="year",columns="month",values="passengers")
print(flights_wide)
sns.heatmap(flights_wide,linewidth=0.5,cmap="YlGnBu")
```
